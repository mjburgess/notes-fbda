### Model Choice : Explanation & Interpretation

```python
df = sns.load_dataset('titanic').dropna()

features = ['age', 'pclass', 'fare']
target = 'survived'

X, y = df[features].values, df[target].values
```

```python

x = df.loc[10, features].values


logreg = LogisticRegression().fit(X, y)

Explainer(
    training_data=X, 
    training_labels=y,
    feature_names=features
).explain_instance(
    x, logreg.predict_proba
).show_in_notebook(show_table=True)
```

Each coefficient in a logistic regression is related to a ratio of log-odds. 

$ log \frac{odds_{survival}}{odds_{death}} = \beta_1x_1 + \beta_2x_2 + \dots $

We can intepret each coefficient ($\beta$) as the size of effect on the odds (of survival) if a unit change is made to a feature whilst holding the others constant. 

Values of $e^\beta$ over 1 will tend to increase probability of survival, whereas values under 1 will decrease it. Below, if I raise age I lower my change of surivial. If I raise my fare I increase it. (And vice versa: a decrease in age, increases my changes). 


```python
dict(zip(features, *np.exp(logreg.coef_)))
```

In the case of a neural network (eg., below) we have many concurrent logistic regressions running over complex interactions of the input... it is generally not possible to ascribe an interpretation to the coefficients of these kinds of complex models. 

```python
from sklearn.neural_network import MLPClassifier 

nn_model = MLPClassifier((15, 2)).fit(X, y)


Explainer(
    training_data=X, 
    training_labels=y,
    feature_names=features
).explain_instance(
    x, nn_model.predict_proba
).show_in_notebook(show_table=True)

```

```python
nn_model.coefs_[0][0, :]
```

## Binary classification trees

* binary trees
    * each node corresponds to a question asked about the data
   
* e.g. fictional task: 
    * distinguish: hawks, penguins, dolphins, and bears
    * series of binary questions that will lead to which animal it might be 
    * i.e. narrow down the outcomes


* with continuous features, you question them via thresholds
    * assigning the majority class in that leaf to all points within a leaf 

* The simplest parameter to regularize a tree is the `maximum_depth`
    * how many "if-else" questions can be asked before deciding which class a sample lies in
    * important to tune for trees and tree-based models

* For a classifier, regularization amounts to limiting the number of partitions of the feature space
    * ie., how many ways it is split

#### thresholding a single feature

* The first line in tree asks: if "X" < decision boundary, for every data point.
    * i.e. **what classification should this data point be assigned**
    * how well does this boundary separate the two classes
    * learning: axis parallel hyperplane
    * 'value' row in each node = number of points assigned to each class
    * 'samples' row in root node is number of points we have trained on (default of train_test_split is 75% into training)
    
    
* how decision boundary is known?
    * **find the best feature and the best threshold to minimize impurity**
        * greatest improvement in purity
        * i.e. tries to make the resulting subsets, the two halves in the above, to be as "pure" as possible 
        * (to consist mostly of one class - i.e. homogeneous wrt the outcome variable)
            * searching over all possible features and all possible thresholds
                * thresholds are all values of the feature that we're observing - iterating over each feature & ask "is this a good way to split the data set into two classes"
                    * i.e. **shapes = squares (at each split, only one feature involved)**

#### repeats, iterates recursively

* find the best feature and the best threshold to minimize impurity, for each of the two classes
    * i.e. model will predict accordingly, further separate the points into the two classes
    
    
* i.e. **recursive partitioning**

### Decision Tress: Measures of impurity

* i.e. method of choosing best split
* we want to quantify the split
    * e.g. "low" if all data in one class (pure node)

* criterion parameter of `DecisionTreeClassifier`:
    * gini: Gini impurity (default)
    * entropy: Information gain
        * expensive to compute

#### Gini Impurity

*  faster, default, preferred

* $H_{gini}(X) = \sum_{k \in classes}(p_{mk} (1 - p_{mk})$

* **probability of incorrectly classifying** a randomly chosen element in the dataset if it were randomly labeled according to the class distribution in the dataset.

* degree of Gini index varies between 0 and 1
    * minimum (zero) when all cases in the node fall into a single target category (all elements belong to a certain class)
    * 1 denotes that the elements are randomly distributed across various classes
    

#### Cross-Entropy

* degree of uncertainty, impurity or disorder
* (over classes) 
* (want to minimise)

$H_{xe}(X) = - \sum_{k \in classes} p_{mk} log(p_{mk})$

## Ensembles 

* combine several weak(er) models to produce a stronger ensemble

### Problems with a Single Model

* eg., a decision tree
    * low bias and high variance 
    * fits well, but predictions sensitive to small permutations in features
    * e.g., 
        * $\hat{y_1} = \hat{f}(x_1 = 5, x_2 = 0.3) = 1$
        * $\hat{y_2} = \hat{f}(x_1 = 4.9, x_2 = 0.31) = 0$
    
### Averaging Models
* eg., RandomForestClassifier

* Base models are trained in independently and averaged
    * averaging models exploits the independence between them
    * the prediction is the average of each of their predictions
    * this reduces the varience of the predictions
    * ie., $ \hat{y} = mean(\hat{f_1}(x) \dots \hat{f_n}(x)) $ 
    * $\hat{y}$ is more stable as variations in each $\hat{f}$ are smoothed out
    

    
#### Bootstrap Aggreation (Bagging)

* Bootstrapping is any statical proces which produces an estimate from multiple random samples
    * a very general purpose method for arriving at low-varience estimates
    * each random sample represents a possible way the data may have been
    * averaging across many such permutations provides better estimates than whole-group measures
    
* Bagging algorithms fit base models to random subsets of the original dataset
    * i.e. each model is exposed to a different subset of data (boostrapping)
    * and then aggregates their individual predictions to form a final prediction
        * either by voting (for classification) or by averaging (for regression)
        * equal weight for each base model


### Aside: Boosting
* eg., the `xgboost` library 

* base models are trained *sequentially* (dependently)
    * so that later models begin from the solution of earlier models
    * their predictions are averaged (so that early-pass and late-pass count the same)
    * this reduces the bias in the predictions
    * ie., the systematic degree to which predictions are out 

* "boosted"
    * each model learns to fix the prediction errors of a prior model
    * weighing previously mislabeled examples with higher weight 
    * reduces bias 
    * may overfit (and thus increase varience)

## Review: Decision trees

#### Advantages

* fast to train
* easy to understand (predicts most common class in leaf, after traverse tree from top to bottom)
- intuitive -> commonly used models both for classification and regression (popular in industry)
- invariant to scaling of the features --> less preprocessing
- distribution of the data is not important 
- easy to handle: categorical variables, missing values
- can build stronger models, by combining multiple trees
- non paramteric (tree model can become more and more flexible, if given more data) (number of free parameters grows with the number of samples)
- non linear
- easy to interpret

#### Limitations
* tend to overfit the training data

#### In Practice
* combine multiple trees to reduce overfitting
    * so models generalize 

* random forests
* gradient boosted trees



# Appendix: Code

```python
def plot_depth(X, y, maxdepth = 1):
    clf = DecisionTreeClassifier(max_depth=maxdepth)
    clf.fit(X, y)
    
    
    plt.subplots(1, 2, figsize=(10, 5))    
    
    # left plot
    plt.title(f"depth = {maxdepth}")
    plt.subplot(1, 2, 1)
    
    eps = X.std() / 2.0/
    x_min, x_max = X[:, 0].min() - eps, X[:, 0].max() + eps
    y_min, y_max = X[:, 1].min() - eps, X[:, 1].max() + eps

    X1, X2 = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    
    y_prob = clf.predict_proba(np.c_[X1.ravel(), X2.ravel()])[:, 1].reshape(X1.shape)
    
    ax = plt.gca()
    ax.contourf(X1, X2, y_prob, levels=[0, 0.5, 1], colors=['lightblue', 'orange'])
    plt.scatter(X[:, 0], X[:, 1], c=y)


    # right plot
    plt.subplot(1, 2, 2) 
    plt.title(f"tree = {maxdepth}")
    plt.tight_layout()
    plot_tree(clf, max_depth = maxdepth, impurity = False, filled = True, label = "root")
```
