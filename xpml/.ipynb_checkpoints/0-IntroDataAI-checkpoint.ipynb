## Buzzwords -- Marketing Speak

* Data

* Data Science
    * vs. Analytics / Business Intelligence
        * Analysis: history, facts, known quantities
        * Informs action by evidencing intution
            * confirm, evidence by confimation
            * reports
    * Science
        * Informing action with clear measures of uncertainty, models
        * Explanation: relationships/models that account for facts
        * Prediction: use of relationships to guess / infer 
        * refutes, evidences by refutation
            * rule out bad ideas
            
* Big Data
    * alert business to the need for alterative skills, tools, etc.
    * hire differnt people, doifferent toosl., 
    
    * `Big data is a field that treats ways to analyze, systematically extract information from, or otherwise deal with data sets that are too large or complex to be dealt with by **traditional data-processing application software**. `
    * The use of non-traditional tools
        * tradition: business-intelligence tooling
            * relational

* Machine Learning
    * Compuational, Statistical, INference
        * Inferring relationships between statistical variables 
            * using computers
        * Stat Variable = Column = Field
            * Variable whose values are obtained from some sample
    * Use Cases
        * Associative-Predictive
            * Associative - no reason to connect input -> output
            * typically: software product            
            * machine should "do the right thing" for *any* reason!
            * eg., loud voice -> fast, always? no.. but who cares 
            
        * Explanatory-Insightful
            * need good reason to suppose input -> ouptut
            * typically: report, insight project, internal focus
            
        * Exploratory 
            * discovering potential patterns
            
    * Generalizes to unseen situations
        * solves problems not explicitly coded
        
    * Examples
        * Automated Warehouses
        * Autonomous Vechicles


* Artifical Intelligence
    * Weak AI 
        * highly specific defintion of problme
        * automation by a toool
        * to perform as well or better than human
        * eg., rock rolling down a hill, or opening a door, beats a child
        
    * Strong AI
        * technqiues/skills/etc. which solve one specific problem in environment A, generalize to other problems in differnet environemetnts 
        * eg., driving in dry weather, can still drive in wet weather
        
    * AI    
    * Neural Networks
        * Deep Learning
