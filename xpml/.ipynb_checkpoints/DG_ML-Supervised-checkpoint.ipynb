## Supervised Learning

So far we have simply said the goal of machine learning is to find a $\hat{y} = \hat{f}(x)$ which approximates a real $f$. 

In the case of supervised learning $\hat{y}$ is a prediction: regression if $y \in \mathbb{R}$ and classification if $y \in \{-1, 0, 1...\}$.

The question remains: *how* do we find the model, $\hat{f}(x)$ ?

Notice that the models above have had *parameters* ($w_0..w_n$). The question of finding a good model, when specifying a "template" in terms of parameters, then just amounts to finding the best value of some parameters. 

The "best value" will be defined according to the total loss incurred on the in-sample (/training) dataset. 

So the supervised learning problem is:

Find:
$\hat{y} = \hat{f}(X; w_0...w_n)$

which minimizes 

$\Sigma \, l(\hat{y}, y) = \Sigma \, l(\hat{f}(X, y)$ 

by varying $w_0..w_n$ 

and computing over the training set 

$\{(X, y), ...)\}$

In english:

Find a relationship ($\hat{f}$) which predicts an unobserved target variable $y$ from observed feature variables ($X$) by varying the parameters of this relationship until it best matches the historical data -- where *best* is defined by a loss, $l(y, \hat{y})$, which compares known targets $y$ to predictions $\hat{y}$ generated by $\hat{f}$. 

### The General Approach to Prediction

There are two steps to solving the prediction problem: 1. finding the best parameters to the model (training, or "learning"); and 2. using the model to predict.

## Quick Example: Linear Regression

We will look at linear regression and linear models in more detail later. However it provides a simple way to illustrate the above formula without much extra complexity. 

In the `train` step we learn the best parameters (using sklearn's `LinearRegression.fit`) and in the `predict` step we use those parameters to calculate a prediction. 

The remaining mystery is how `.fit` works, which will be considered in the "Optimization" section at the end. For now, consider the workflow: fit then predict. 

(NB. you can imagine `.fit` merely trying a range of sensible values for the parameters, improving each guess by considering how increasing/decreasing the parameter affects the total loss, ie., finding the best parameters such that the total loss over the entire training set ($loss(\hat{f}(X; w_0...), y)$ is minimized.)

```python
def train_linear(X, y):
    model = LinearRegression().fit(X, y)
    parameters = model.coef_[0], model.intercept_
    return parameters

def predict(model, x):
    w1, w2 = model
    return w1 * x + w2

model = train_linear([[1], [2], [3]], [1.1, 2.2, 3.3])
yhat = predict(model, 4)
```

## Nearest Neighbours

The Nearest Neighbour algorithm is one of the simplest classification solutions. 

The training step is simple: just remember the entire historical training data set. 

In the prediction step, the model is just to find the "most similar" historical $x$ and use its $y$ to make the prediction.

The idea is straightforward: to predict a $\hat{y}$ given an $x$ we find the closest match in the database, and use its known $y$. 

Eg., to predict whether a teenager will crash a car ($\hat{y}$) on their first drive, given their age $x$, we take a historical database of teenagers and find a teenager with the closest age, and report *their* crash as our prediction. 

This, of course, isn't likely to be very accurate in most cases. In the k-Nearest Neighbours generalization, we find $k$ most similar observations and let them vote on the $y$ -- eg., if 2/3 are +1 and 1/3 are -1 then the prediction is +1. 

```python
x_age = np.linspace(16, 19, 5)
y_crash = (x_age <= 17.5).astype(int)

print(x_age)
print(y_crash)
```

If I find the parameter is 17, what is the error/distance for each point...

```python
[distance(17, xt) for xt in x_age]
```

```python
y_index = np.argmin(distance(17, xt) for xt in x_age)

y_crash[y_index]
```

Generalizing this, we can describe the training step as returning a function, `trained_model` which accepts an $x$ and returns the $y$ for the nearest historical $x$.  

```python
from scipy.spatial.distance import euclidean as distance 

# training step
x_train = x_age
y_train = y_crash

def fhat_knn(x_unseen):
    alldist = [distance(x_unseen, x_historical) for x_historical in x_train]
    return y_train[ np.argmin(alldist) ]

# predicting step
fhat_knn(19)
```

### Aside: using currying to set parameters

In the following code you can read `x_train` and `y_train` as default parameters for the function `train_model`. 

This technique is called currying and allows you to dynamically adapt the value of defaults. 

With this system its easier to generate multiple different knn models (ie., changing $k$).

```python
def train_knn(x_train, y_train):
    def trained_model(x):
        return y_train[ 
            np.argmin([distance(x, xt) for xt in x_train]) 
        ]
    
    return trained_model 
```

Here $\hat{f}$ is the function returned by `train_knn` -- the name `fhat` is less descriptive than either `model` or `predict` etc. but is used here just to illustrate correspondance with the notation used above. 

```python
fhat = train_knn(x_age, y_crash)
yhat = fhat(20)
```

```python
terms = [ "not crash","crash"]

print("A 20 year old will " + terms[yhat]) 
```

### k-Nearest

We can use `np.argsort` to rank all values in the distance list... (and hence return the bottom `k`)

```python
from statistics import *
```

```python
index_ranks = np.array([
    distance(17, xt) for xt in x_age
]).argsort()

print(index_ranks[:3])
print(x_ages[index_ranks[:3]])
```

```python
def outer(x=45):
    def inner():
        return x
    return inner
    
f = outer()
f()
```

```python
def train_knn(x_train, y_train, k):
    def trained_model(x):
        nearest = np.array([
            distance(x, xt) for xt in x_train
        ]).argsort()[:k]
        
        return mode(y_train[nearest])
    
    return trained_model 
```

```python
model = train_knn(x_age, y_crash, 3)
model(18)
```

Suppose I choose `k = 5`, note that this is the length of `y_crash`, therefore to predict for any $x$, this choice of $k$ will consider the entire dataset and so will always predict `mode(y_crash)`.

```python
predict = train_knn(x_age, y_crash, 5)

predict(20), mode(y_crash)
```

For a smaller $k$, the algorithm will choose a $k$-sized subset of `y_crash` which is the "closest" given the proximity of `x` to the accompanying `x_train`.

## Comparing Models

* recall, that the loss is the way we find a parameter
    * how good are some parameters(+model) at estimatnig a historical dataset

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
```

```python
data = load_breast_cancer()
X = data['data']
y = data['target']
```

```python
def score_cancer(y, yhat):
    # malignant = 0, benign = 1
    confusion_marix = np.array([
        [0 , 1],  # 00, 01 # y yhat
        [7 , 0]   # 10, 11 # y yhat
    ])
    return sum([ confusion_marix[y[i], yhat[i]] for i in range(len(y)) ])
```

```python
score_cancer(y, KNeighborsClassifier(1).fit(X, y).predict(X))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
(score_cancer(y_test, KNeighborsClassifier(1).fit(X_train, y_train).predict(X_test)),
score_cancer(y_test, KNeighborsClassifier(3).fit(X_train, y_train).predict(X_test)),
score_cancer(y_test, KNeighborsClassifier(11).fit(X_train, y_train).predict(X_test)))
```

```python
results = {}
for k in range(1, 41):
    model = KNeighborsClassifier(k).fit(X_train, y_train)
    results[k] = score_cancer(y_test, model.predict(X_test))
    
sns.lineplot(list(results.keys()), list(results.values()));
```

```python
print(classification_report(y_test, KNeighborsClassifier(21).fit(X_train, y_train).predict(X_test)))
```

```python
N, P = -1, +1
print(classification_report( [N, P, N, P, P, P, P], [N, N, P, P, N, P, P], target_names=['Negtv', 'Postv'] ))
```

## Review: The Machine Learning Workflow

In the code below the key stages of the machine learning workflow are partitioned into a clear API -- the name of the stage is given by the name of a function, and its requirements are given by its parameters. 



```python
def numpy_split(X, y, percent=20):
    assert len(X) == len(y)
    
    selected = np.random.randint(0, len(X), len(X) // percent)
    
    X_a, y_a = X[selected, :], y[selected]
    X_b, y_b = np.delete(X, selected, 0), np.delete(y, selected, 0)
    
    return (X_a, y_a, X_b, y_b)
 
```

```python
def select_Xy(dataframe, features, target):
    return dataframe[features].values, dataframe[target].values

def test_split(X, y):
    return numpy_split(X, y)

def train(model, X_train, y_train):
    return model.fit(X_train, y_train)

def validation_loss(trained_model, X_val, y_val):
    return 1 - trained_model.score(X_val, y_val)
    
def validate(models, X_train, y_train):
    X_val, y_val, X_tr, y_tr = numpy_split(X_train, y_train)
    return min(
        (validation_loss(train(m, X_tr, y_tr), X_val, y_val), m) 
        for m in models
    )

def evaluate(model, X_test, y_test):
    return model.score(X_test, y_test)


def predict(model, x):
    return model.predict(x)

```

```python
data = sns.load_dataset('tips')
data.sample(3)
```

```python
print("A Sample ML Workflow")
print()

features = ['tip', 'size']
target = 'total_bill'

X, y = select_Xy(data, features, target)
print(f"Xy Selection Complete!\t X = {features}, y = {target}")

X_test, y_test, X_train, y_train = test_split(X, y)
print("Test Split Complete!\t Train-Test Split at 20%")

models = [LinearRegression(), Ridge(), Lasso()]
print("Model Choice Complete!\t Choosing models: ", *[
    type(m).__name__ for m in models
])

model_loss, model_best = validate(models, X_train, y_train)
print(f"Validation Complete!\t Best model was {type(model_best).__name__} with total loss {model_loss : .2f}")

model_score = evaluate(model_best, X_test, y_test)
print(f"Evaluation Complete!\t Model performance on test set: {model_score:.2f}")

model = train(model_best, X, y)
print("Retraining Complete!\t Model retrained on entire dataset")

x_unseen = (2.00, 2)
yhat = predict(model, [x_unseen])
print(f"Deployment Complete!\t Model prediction for x = {x_unseen} is {yhat[0] : .2f}")
```

Above we looked at the workflow by spelling out how each step will work manually, and often the choices weren't perfect to keep the implementation simple.

For example, the validation set is only split once into 20/80 and we'd prefer to use cross-validation with *five* such splits. 

In the code below, the key elements of the sklearn API are used -- as in exactly the *workflow* above -- but now with each step a little more optimial.

When contrasting the above with the below, note that the outline above partitions each step into reusable functions which can be extracted into an external library (rather than being embdeded in a notebook). This style of programming is closer to how a real-world deployment of a project should look.
